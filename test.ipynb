```python
import pandas as pd
import json



# Ruta del archivo .jsonl
file_path = ""

# Lista para almacenar los datos normalizados
data = []

# Leer y desanidar cada l√≠nea
with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        record = json.loads(line)
        normalized = pd.json_normalize(record)
        data.append(normalized)

# Concatenar todas las filas en un solo DataFrame
df = pd.concat(data, ignore_index=True)

# Mostrar las primeras columnas como ejemplo
print(df.columns.tolist())
print(df.head())
```
